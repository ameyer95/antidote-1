```python
import pandas as pd
import numpy as np
from datetime import timedelta
from sklearn.model_selection import train_test_split
```

### Load raw data
We are using the FAIR ML version (https://www.kaggle.com/danofer/compass-fairml-getting-started) to avoid propogating the data with people's names, etc., and to allow for better comparison with other fairness-focused work.

```python
fair_data = pd.read_csv("../data/raw_data/propublica_data_for_fairml.csv", index_col=False)
```

```python
# use default train/test split size (0.75/0.25)
# unfortunatly, forgot a random seed- but the exact split we used is in the ../data directory.
fair_data_train, fair_data_test = train_test_split(fair_data)
fair_train = pd.DataFrame(fair_data_train)
fair_test = pd.DataFrame(fair_data_test)
```

```python
fair_train.to_csv("../data/compas_fairml_train.data",",",index=False, header=False)
fair_test.to_csv("../data/compas_fairml_test.data", ",", index=False, header=False)
```
